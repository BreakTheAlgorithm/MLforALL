## Chapter 5 — Intro to ML | Chapter 6 — Data Preprocessing

| Topic | Formula / Key Point |
|-------|--------------------|
| Supervised vs Unsupervised | Supervised has labels (y); unsupervised does not |
| Bias-variance tradeoff | Total error = Bias² + Variance + Irreducible noise |
| StandardScaler | $z = (x - \mu) / \sigma$ — zero mean, unit variance |
| MinMaxScaler | $x' = (x - x_{min}) / (x_{max} - x_{min})$ — scales to [0,1] |
| One-Hot Encoding | Converts categorical to binary columns; drop first to avoid multicollinearity |
| Missing values | Numeric: impute with median; Categorical: impute with mode or 'Unknown' |
| Feature selection | Correlation filter, mutual info, RFECV, L1 regularisation |
| Pipeline | `Pipeline([('scaler', StandardScaler()), ('clf', LogisticRegression())])` |

## Chapter 7 — Regression | Chapter 8 — Classification

| Algorithm | Hyperparameters | Best For |
|-----------|----------------|----------|
| Linear Regression | `fit_intercept` | Continuous target, linear relationship |
| Ridge (L2) | `alpha` | Correlated features, all features matter |
| Lasso (L1) | `alpha` | Feature selection, sparse models |
| Logistic Regression | `C`, `penalty` | Binary/multiclass, interpretable |
| Decision Tree | `max_depth`, `min_samples_split` | Non-linear, interpretable rules |
| Random Forest | `n_estimators`, `max_features` | Robust, handles missing, fast |

**Key Metrics:**

$$\text{MSE} = \frac{1}{n}\sum(y_i - \hat{y}_i)^2 \quad \text{RMSE} = \sqrt{\text{MSE}} \quad R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$$

$$\text{Precision} = \frac{TP}{TP+FP} \quad \text{Recall} = \frac{TP}{TP+FN} \quad F_1 = \frac{2 \cdot P \cdot R}{P+R}$$

## Chapter 9 — Model Evaluation | Chapter 10 — Ensemble Methods

| Technique | Purpose | Code |
|-----------|---------|------|
| k-Fold CV | Robust performance estimate | `cross_val_score(model, X, y, cv=5)` |
| GridSearchCV | Exhaustive hyperparameter search | `GridSearchCV(model, param_grid, cv=5)` |
| Bagging | Reduces variance | `BaggingClassifier` |
| Random Forest | Bagging + feature subsampling | `RandomForestClassifier` |
| AdaBoost | Sequential reweighting of errors | `AdaBoostClassifier` |
| Gradient Boosting | Fits residuals sequentially | `GradientBoostingClassifier` |
| XGBoost / LightGBM | Fast, regularised GBM | `XGBClassifier`, `LGBMClassifier` |
| Stacking | Meta-model on OOF predictions | `StackingClassifier` |

> 📌 **Rule of thumb:** Ensemble methods almost always outperform single models on tabular data. GBM variants win most Kaggle tabular competitions.

## Chapter 13 — NLP | Chapter 14 — Neural Networks | Chapter 15 — Deep Learning

**NLP Pipeline:**
```
Text → Tokenise → (Remove stopwords) → Vectorise (BoW/TF-IDF/Embeddings) → Model
```

$$\text{TF-IDF}(t,d) = \text{TF}(t,d) \times \log\frac{N}{df(t)}$$

**Neural Network Forward Pass:**
$$a^{(l)} = \sigma(W^{(l)} a^{(l-1)} + b^{(l)})$$

| Activation | Formula | Use Case |
|-----------|---------|----------|
| ReLU | $\max(0, x)$ | Hidden layers (default) |
| Sigmoid | $1/(1+e^{-x})$ | Binary output |
| Softmax | $e^{x_i}/\sum e^{x_j}$ | Multiclass output |
| tanh | $(e^x - e^{-x})/(e^x + e^{-x})$ | Older RNNs |

## Chapter 16–19 — End-to-End, Deployment, Ethics, Time Series

| Topic | Key Takeaway |
|-------|-------------|
| ColumnTransformer | Apply different preprocessing to numeric vs categorical columns |
| Pipeline | Chain preprocessing + model; prevents data leakage in CV |
| Threshold tuning | `predict_proba()` + custom threshold to trade precision for recall |
| joblib.dump/load | Save/load sklearn models to disk |
| Predict API | Validate inputs before calling model; return structured dict |
| Model drift | Monitor input distribution + output distribution over time |
| Fairness metrics | Equal accuracy, equalised odds, demographic parity |
| Stationarity | ADF test; differencing to remove trend before ARIMA |
| Lag features | Turn time series into tabular problem for any ML model |
| TimeSeriesSplit | CV without data leakage for temporal data |

## 📐 Formula Reference Card

| Metric | Formula |
|--------|--------|
| MSE | $\frac{1}{n}\sum(y_i-\hat{y}_i)^2$ |
| RMSE | $\sqrt{\text{MSE}}$ |
| MAE | $\frac{1}{n}\sum|y_i-\hat{y}_i|$ |
| R² | $1 - \frac{\sum(y_i-\hat{y}_i)^2}{\sum(y_i-\bar{y})^2}$ |
| Precision | $TP/(TP+FP)$ |
| Recall | $TP/(TP+FN)$ |
| F1 | $2PR/(P+R)$ |
| Log Loss | $-\frac{1}{n}\sum[y\log\hat{p}+(1-y)\log(1-\hat{p})]$ |
| L1 (Lasso) | $\text{Loss} + \alpha\sum|w_i|$ |
| L2 (Ridge) | $\text{Loss} + \alpha\sum w_i^2$ |
| Entropy | $H = -\sum p_i \log_2 p_i$ |
| Gini | $1 - \sum p_i^2$ |
