## 🐍 Option 2 — Anaconda (Recommended for Local Setup)

Anaconda bundles Python, Jupyter, and 250+ scientific packages.

**Install Anaconda:**
1. Download from [anaconda.com/download](https://www.anaconda.com/download)
2. Choose Python 3.10+ installer for your OS
3. Run installer (accept defaults, add to PATH on Windows)

**Create a dedicated environment for this book:**
```bash
conda create -n mlbook python=3.10
conda activate mlbook
conda install numpy pandas matplotlib seaborn scikit-learn jupyter
pip install statsmodels xgboost lightgbm shap
jupyter lab
```

**Why use conda environments?**
- Isolate project dependencies
- Avoid version conflicts between projects
- Easy to recreate with `conda env export > environment.yml`

## 🔧 Option 3 — pip + venv (Lightweight)

For users who just want Python without Anaconda:

```bash
# Install Python 3.10+ from python.org

# Create virtual environment
python -m venv mlenv

# Activate (Linux/Mac)
source mlenv/bin/activate

# Activate (Windows)
mlenv\Scripts\activate

# Install packages
pip install numpy pandas matplotlib seaborn scikit-learn
pip install jupyterlab statsmodels xgboost lightgbm tensorflow shap

# Launch Jupyter
jupyter lab
```

**Save your environment:**
```bash
pip freeze > requirements.txt
pip install -r requirements.txt  # recreate on another machine
```

## 🔥 TensorFlow / PyTorch Setup

**TensorFlow (Chapter 15):**
```bash
pip install tensorflow          # CPU-only (works for this book)
pip install tensorflow-gpu      # if you have an NVIDIA GPU
```

Verify:
```python
import tensorflow as tf
print(tf.__version__)           # Should be 2.x
print(tf.config.list_physical_devices('GPU'))  # [] means CPU mode
```

**PyTorch (alternative):**
```bash
pip install torch torchvision   # CPU build
# For GPU, visit pytorch.org/get-started for CUDA-specific command
```

````python
# ─────────────────────────────────────────────────────────────
# Verify your environment — run this cell to check all packages
# ─────────────────────────────────────────────────────────────
import sys

checks = [
    ('Python',      sys.version.split()[0]),
]

packages = [
    'numpy', 'pandas', 'matplotlib', 'seaborn',
    'sklearn', 'statsmodels', 'xgboost', 'lightgbm',
    'joblib', 'scipy'
]

for pkg in packages:
    try:
        m = __import__(pkg)
        ver = getattr(m, '__version__', 'OK')
        checks.append((pkg, ver))
    except ImportError:
        checks.append((pkg, '❌ NOT INSTALLED'))

# Check TF separately
try:
    import tensorflow as tf
    checks.append(('tensorflow', tf.__version__))
except ImportError:
    checks.append(('tensorflow', '⚠️ not installed (needed for Ch15)'))

print('Environment Check:')
print('-'*40)
for name, ver in checks:
    status = '✅' if '❌' not in str(ver) and '⚠️' not in str(ver) else ''
    print(f'{name:<20} {ver} {status}')
````
