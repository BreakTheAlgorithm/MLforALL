## 📖 Section A — Variables, Data Types & F-strings

Python has four core scalar types: `int`, `float`, `str`, `bool`.

```python
num_features   = 12        # int
model_accuracy = 0.934     # float
model_name     = 'XGBoost' # str

```

**F-strings** (Python 3.6+) are the cleanest way to format output:
```python
print(f'Model: {model_name} | Accuracy: {model_accuracy:.1%} | Features: {num_features}')
# → Model: XGBoost | Accuracy: 93.4% | Features: 12
```

> 💡 **Key Rule**
> `f'text {variable:.2f} more text'`
> The `:` after the variable name starts a format spec — `.2f` = 2 decimal places float, `.1%` = percentage with 1dp.

````python
# ─────────────────────────────────────────────────────────────
# Demo: Data Types and F-strings
# ─────────────────────────────────────────────────────────────

# Core ML model metadata
num_features   = 12
model_accuracy = 0.934
model_name     = 'XGBoost'
is_converged   = True

print(f'Model      : {model_name}')
print(f'Accuracy   : {model_accuracy:.1%}')
print(f'Features   : {num_features}')
print(f'Converged  : {is_converged}')
print(f'Type check : {type(num_features)} | {type(model_accuracy)} | {type(model_name)} | {type(is_converged)}')

# Salary prediction demo (key pattern from the book)
print('\n--- Salary Prediction Demo ---')
def predict_salary(years_experience, base=300000, increment=150000):
    predicted = base + (years_experience * increment)
    return predicted

for years in [0, 1, 2, 3, 5, 7, 10]:
    print(f'{years:2d} yrs → ₹{predict_salary(years):>10,}')
````

## 📖 Section B — Lists, Loops & Comprehensions

Lists are ordered, mutable sequences. They're the Python workhorse before you graduate to NumPy arrays.

| Operation | Syntax | Notes |
|-----------|--------|-------|
| Create | `lst = [1, 2, 3]` | |
| Index | `lst[0]` → 1 | 0-indexed |
| Slice | `lst[1:3]` → [2,3] | end exclusive |
| Comprehension | `[x*2 for x in lst]` | creates new list |
| Enumerate | `for i, v in enumerate(lst)` | gives index + value |

> 💡 **Key Rule**
> `[expression for item in iterable if condition]`
> This is the list comprehension pattern. Use it when transforming or filtering a list in one clean line.

````python
# ─────────────────────────────────────────────────────────────
# Demo: Lists, Slicing, and Comprehensions
# ─────────────────────────────────────────────────────────────

# Zomato restaurant ratings (out of 5)
ratings = [4.2, 3.8, 4.5, 2.9, 4.1, 3.5, 4.8, 2.1, 4.0, 3.9]

print('=== List Operations ===')
print(f'All ratings    : {ratings}')
print(f'First rating   : {ratings[0]}')
print(f'Last rating    : {ratings[-1]}')
print(f'Top 3 (sliced) : {sorted(ratings, reverse=True)[:3]}')

print('\n=== Loop with enumerate ===')
for i, rating in enumerate(ratings[:5], start=1):
    stars = '⭐' * int(rating)
    print(f'Restaurant {i}: {rating} {stars}')

print('\n=== List Comprehensions ===')
good_ratings = [r for r in ratings if r >= 4.0]
print(f'Restaurants with 4.0+ rating: {good_ratings}')
print(f'Count: {len(good_ratings)} out of {len(ratings)}')

average = sum(ratings) / len(ratings)
print(f'\nAverage rating: {average:.2f}')
````

## 📖 Section C — Functions

Functions are the core abstraction in Python. Every well-written ML pipeline is a sequence of functions.

```python
def function_name(param1, param2, default_param=value):
    """One-line docstring describing what this does."""
    result = # ... computation ...
    return result
```

> 💡 **Key Rule**
> Always include a docstring. Always use `return`. Avoid modifying variables outside the function (side effects create bugs).

````python
# ─────────────────────────────────────────────────────────────
# Demo: Functions with defaults, docstrings, and real ML use
# ─────────────────────────────────────────────────────────────

def min_max_normalise(values, new_min=0.0, new_max=1.0):
    """Normalise a list of values to [new_min, new_max] range."""
    v_min = min(values)
    v_max = max(values)
    return [
        new_min + (v - v_min) / (v_max - v_min) * (new_max - new_min)
        for v in values
    ]

swiggy_prices = [199, 350, 89, 799, 449, 129, 599]
normalised = min_max_normalise(swiggy_prices)

print('Swiggy order prices → normalised to [0, 1]:')
for original, norm in zip(swiggy_prices, normalised):
    bar = '█' * int(norm * 20)
    print(f'  ₹{original:4d}  →  {norm:.2f}  {bar}')

print(f'\nMin after normalise: {min(normalised):.2f}  (should be 0.00)')
print(f'Max after normalise: {max(normalised):.2f}  (should be 1.00)')
````

````python
# @title ✅ Solution — Exercise 5.5 (click to expand)
# @markdown > Run this cell to reveal the solution

def categorise_salary(amount):
    if amount < 400000:    return 'Entry'
    elif amount <= 800000: return 'Mid'
    elif amount <= 1500000:return 'Senior'
    else:                  return 'Principal'

raw_data = [
    {'age': 25, 'income': '450000', 'city': 'Mumbai'},
    {'age': 31, 'income': '950000', 'city': 'Bangalore'},
    {'age': 22, 'income': '300000', 'city': 'Pune'},
    {'age': 40, 'income': '1800000','city': 'Delhi'},
    {'age': 28, 'income': '700000', 'city': 'Hyderabad'},
]

def preprocess_features(raw_data):
    """Clean and enrich raw employee records for ML use.

    - Converts income from string to int
    - Adds income_bracket using categorise_salary()
    Returns a new list — does not mutate the original.
    """
    cleaned = []
    for record in raw_data:
        # Build a new dict rather than mutating the original
        clean_record = {
            'age':            record['age'],
            'income':         int(record['income']),   # str → int
            'city':           record['city'],
            'income_bracket': categorise_salary(int(record['income']))  # computed feature
        }
        cleaned.append(clean_record)
    return cleaned

cleaned = preprocess_features(raw_data)
print('Preprocessed Features:')
for record in cleaned:
    print(f'  {record}')

assert all(isinstance(r['income'], int) for r in cleaned)
assert all('income_bracket' in r for r in cleaned)
assert cleaned[0]['income_bracket'] == 'Mid'
assert cleaned[3]['income_bracket'] == 'Principal'
print('\n✅ This pattern — loading, cleaning, and enriching raw dicts — is the foundation of every ML ETL pipeline.')
````

````python
# @title ✅ Solution — Exercise 5.6 (click to expand)
# @markdown > Run this cell to reveal the solution

def average(numbers):
    """Return the mean of a list of numbers."""
    return sum(numbers) / len(numbers)

def even_numbers(numbers):
    """Return only the even numbers from a list."""
    return [n for n in numbers if n % 2 == 0]

# Verify with asserts
assert average([2, 4, 6, 8]) == 5.0,    'Average of [2,4,6,8] should be 5.0'
assert even_numbers([1,2,3,4,5,6]) == [2,4,6], 'Even numbers of [1..6] should be [2,4,6]'
assert even_numbers([1,3,5]) == [],     'No evens in [1,3,5]'

print(f'average([10, 20, 30]) = {average([10, 20, 30])}')
print(f'even_numbers([1..10]) = {even_numbers(range(1,11))}')
print('\n✅ Readiness check passed — you are ready for NumPy!')
````

## 🎤 Interview Q&A

<details>
<summary><strong>Q1: What is the difference between a list and a tuple in Python?</strong></summary>

**Answer:** Both store ordered sequences, but **lists are mutable** (can be changed after creation) while **tuples are immutable** (cannot be changed). Tuples are faster and used for fixed data like coordinates or RGB colours. Lists are used when data changes. In ML, dataset rows are often tuples; model parameters are stored in mutable structures.
</details>

<details>
<summary><strong>Q2: Explain list comprehensions — when should you use them vs a for loop?</strong></summary>

**Answer:** A list comprehension is a concise way to create a list: `[expr for item in iterable if cond]`. Use it for simple transformations and filters in one line. Use a regular for loop when: the logic is complex (multiple conditions, multiple operations per item), or you need to append to an existing list, or readability suffers. Rule of thumb: if it fits in one readable line, use a comprehension.
</details>

<details>
<summary><strong>Q3: What is a dictionary and what is its time complexity for lookup?</strong></summary>

**Answer:** A dictionary is a hash map — it stores key-value pairs where keys must be hashable (usually strings or numbers). Lookup by key is **O(1) average** because Python hashes the key to find the memory location directly, without scanning. This is why dicts are preferred over lists when you need to look up by name rather than position.
</details>

<details>
<summary><strong>Q4: What do *args and **kwargs mean?</strong></summary>

**Answer:** `*args` collects extra positional arguments into a tuple. `**kwargs` collects extra keyword arguments into a dictionary. They allow functions to accept variable numbers of arguments — common in ML libraries like sklearn where `fit(X, y, **params)` passes through extra config options.
</details>

<details>
<summary><strong>Q5: What is the difference between .append() and .extend()?</strong></summary>

**Answer:** `.append(x)` adds a single element to the end of a list (even if `x` is itself a list — it becomes a nested list). `.extend(iterable)` unpacks the iterable and adds each element individually. Example: `[1,2].append([3,4])` → `[1,2,[3,4]]`. `[1,2].extend([3,4])` → `[1,2,3,4]`.
</details>
