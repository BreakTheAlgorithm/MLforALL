## 📖 Section A — What Is Machine Learning?

Traditional programming tells the computer **exactly what rules to follow**. Machine Learning is different: we give the computer **examples** and let it figure out the rules itself.

| Approach | Rule Source | Example |
|---|---|---|
| Rule-Based | Human writes rules | `if spam_words > 3: label = 'spam'` |
| Machine Learning | Machine learns rules from data | Feed 10,000 labelled emails → model learns patterns |

> 💡 **Key Idea**
> `ML = Pattern Recognition from Examples`
> The machine finds statistical relationships in data so you don't have to hardcode them.

```python
# ─────────────────────────────────────────────────────────────
# Demo: Rule-Based vs ML Approach
# Shows the conceptual difference using a spam detection example
# ─────────────────────────────────────────────────────────────

# Rule-based approach (hardcoded logic)
def rule_based_spam(email_text):
    spam_words = ['free', 'winner', 'prize', 'click here', 'limited offer']
    count = sum(1 for word in spam_words if word.lower() in email_text.lower())
    return 'SPAM' if count >= 2 else 'NOT SPAM'

# ML approach (model learns from data — represented here conceptually)
training_emails = [
    ('Congratulations! You are a WINNER — click here for your free prize!', 1),
    ('Meeting rescheduled to 3pm tomorrow', 0),
    ('FREE limited offer — claim your prize now!', 1),
    ('Your invoice from Swiggy is attached', 0),
    ('Win BIG — free entry, click here immediately!', 1),
]

# Rule-based results
print('=== Rule-Based Classifier ===')
for email, label in training_emails:
    result = rule_based_spam(email)
    actual = 'SPAM' if label == 1 else 'NOT SPAM'
    match = '✅' if result == actual else '❌'
    print(f'{match} Predicted: {result:10s} | Actual: {actual:10s} | Email: {email[:50]}...')

print('\nKey Insight: Rule-based works for simple cases but fails when spam evolves!')
```

## 📖 Section B — The Three Types of Machine Learning

### 1. Supervised Learning
**You provide labelled examples** — the model learns to map inputs to outputs.
- Example: Predict Swiggy delivery time given distance, restaurant type, weather
- Tasks: Classification, Regression

### 2. Unsupervised Learning
**No labels provided** — the model finds hidden structure in the data.
- Example: Grouping Flipkart customers by purchasing behaviour
- Tasks: Clustering, Dimensionality Reduction

### 3. Reinforcement Learning
**An agent learns by trial and error**, maximising a reward signal.
- Example: Training a bot to play a game, self-driving car
- Tasks: Game playing, Robotics, Recommendation

> 💡 **Quick Rule**
> `Do your examples have labels? → Supervised  |  No labels? → Unsupervised  |  Reward signal? → Reinforcement`

```python
# ─────────────────────────────────────────────────────────────
# Demo: Visualising ML types with simple data
# ─────────────────────────────────────────────────────────────

np.random.seed(42)
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle('Three Types of Machine Learning', fontsize=14, fontweight='bold')

# Supervised: Labelled data with decision boundary
class0 = np.random.randn(30, 2) - 1.5
class1 = np.random.randn(30, 2) + 1.5
axes[0].scatter(class0[:,0], class0[:,1], c='blue', label='Class A', alpha=0.6)
axes[0].scatter(class1[:,0], class1[:,1], c='red', label='Class B', alpha=0.6)
axes[0].axline((0,0), slope=1, color='green', linestyle='--', label='Decision boundary')
axes[0].set_title('Supervised Learning\n(Labelled data)', fontweight='bold')
axes[0].legend(fontsize=8)

# Unsupervised: Unlabelled data, clusters discovered
c1 = np.random.randn(25, 2) + np.array([-2, -2])
c2 = np.random.randn(25, 2) + np.array([2, 2])
c3 = np.random.randn(25, 2) + np.array([2, -2])
all_pts = np.vstack([c1, c2, c3])
colors = ['purple']*25 + ['orange']*25 + ['teal']*25
axes[1].scatter(all_pts[:,0], all_pts[:,1], c=colors, alpha=0.6)
axes[1].set_title('Unsupervised Learning\n(Clusters discovered from unlabelled data)', fontweight='bold')

# Reinforcement: Agent, environment, reward
axes[2].text(0.5, 0.6, '🤖 Agent', ha='center', fontsize=16, transform=axes[2].transAxes)
axes[2].text(0.5, 0.4, '↕ Action / Reward', ha='center', fontsize=10, transform=axes[2].transAxes, color='green')
axes[2].text(0.5, 0.2, '🌍 Environment', ha='center', fontsize=16, transform=axes[2].transAxes)
axes[2].set_title('Reinforcement Learning\n(Agent learns from reward signal)', fontweight='bold')
axes[2].axis('off')

plt.tight_layout()
plt.show()
print('The three paradigms — each suited to a different type of problem.')
```

## 📖 Section C — The Standard ML Workflow (9 Steps)

Every ML project — from spam detection to Zomato delivery time prediction — follows the same pipeline:

| Step | Name | What You Do |
|------|------|-------------|
| 1 | Define the Problem | What do you want to predict? What type of ML? |
| 2 | Collect Data | Gather raw examples (CSV, API, scraping) |
| 3 | Explore Data (EDA) | Understand distributions, spot outliers |
| 4 | Preprocess Data | Clean, fill missing values, encode categories |
| 5 | Feature Engineering | Create new useful features from existing ones |
| 6 | Select Model | Choose algorithm (linear, tree, neural net…) |
| 7 | Train Model | Fit model to training data |
| 8 | Evaluate Model | Test on held-out data, compute metrics |
| 9 | Deploy & Monitor | Ship to production, watch for drift |

```python
# @title ✅ Solution — Exercise 4.2 (click to expand)
# @markdown > Run this cell to reveal the solution

def data_quality_report(data):
    report = {
        'missing_values': {},
        'type_issues': [],
        'workflow_steps': {}
    }

    # Check for None / 'N/A' missing values
    for col in ['age', 'income', 'city', 'label']:
        count = sum(1 for row in data if row.get(col) is None or row.get(col) == 'N/A')
        if count > 0:
            report['missing_values'][col] = count

    # Check for type mismatches (income should be numeric)
    for i, row in enumerate(data):
        if row['income'] not in (None, 'N/A'):
            try:
                int(row['income'])
            except (ValueError, TypeError):
                report['type_issues'].append(('income', i, f"Non-numeric value: {row['income']}"))

    # Map issue types to ML workflow steps
    report['workflow_steps'] = {
        'missing_values':  'Step 4 — Data Preprocessing (fill or drop missing values)',
        'type_issues':     'Step 4 — Data Preprocessing (type coercion/cleaning)',
        'inconsistent':    'Step 3 — Exploratory Data Analysis (detect anomalies)',
    }

    return report

report = data_quality_report(raw_data)
print('Data Quality Report:')
for section, content in report.items():
    print(f'  {section}: {content}')

assert 'missing_values' in report
assert 'workflow_steps' in report
print('\n✅ Solution verified! This is Step 4 of the ML workflow — a critical real-world skill.')
```

```python
# @title ✅ Solution — Exercise 4.3 (click to expand)
# @markdown > Run this cell to reveal the solution

import math

def euclidean_distance(point1, point2):
    # Unpack the two coordinates for each point
    x1, y1 = point1
    x2, y2 = point2
    # Apply the Euclidean formula: sqrt of sum of squared differences
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def predict_1nn(training_data, query_point):
    min_dist  = float('inf')   # Start with "infinite" distance
    best_label = None

    for x, y, label in training_data:
        # Compute distance from query point to this training point
        dist = euclidean_distance(query_point, (x, y))
        # Keep track of the closest point found so far
        if dist < min_dist:
            min_dist   = dist
            best_label = label

    return best_label

# Verify
test_cases = [
    ((1.2, 2.1), 'Junior'),
    ((5.8, 6.2), 'Senior'),
    ((1.8, 2.3), 'Junior'),
]
print('1-NN Classification Results:')
for query, expected in test_cases:
    predicted = predict_1nn(training_data, query)
    print(f'✅ Query: {query} | Predicted: {predicted}')
    assert predicted == expected

print('\n✅ This is exactly what sklearn KNeighborsClassifier(n_neighbors=1) does internally!')
print('   sklearn just adds: multiple neighbours, efficient k-d tree search, and cross-validation support.')
```

## 🎤 Interview Q&A

<details>
<summary><strong>Q1: What is the difference between supervised and unsupervised learning?</strong></summary>

**Answer:**
In **supervised learning**, every training example has a label (the correct answer). The model learns to map inputs → labels. Examples: spam classification, salary prediction, image recognition.

In **unsupervised learning**, there are no labels. The model discovers hidden structure — clusters, patterns, or compressed representations — without being told what to look for. Examples: customer segmentation, topic modelling, anomaly detection.

**Quick way to remember:** Supervised = you provide the teacher's answer key. Unsupervised = no answer key, the model groups things itself.
</details>

<details>
<summary><strong>Q2: What is the bias-variance tradeoff?</strong></summary>

**Answer:**
- **Bias**: Error from wrong assumptions (model too simple → underfits → misses patterns)
- **Variance**: Error from sensitivity to training data (model too complex → overfits → memorises noise)

High bias → underfitting (e.g., linear model for a curved relationship)
High variance → overfitting (e.g., 1000-node decision tree on 100 samples)

The tradeoff: reducing one tends to increase the other. The goal is to find the sweet spot.
</details>

<details>
<summary><strong>Q3: Why do we split data into train and test sets?</strong></summary>

**Answer:**
Because we want to measure how well the model **generalises** to new, unseen data — not how well it memorised the training examples.

If we evaluate on training data, an overfit model will score 100% but perform terribly in production. The test set simulates production data.

Standard splits: 80/20 or 70/30. For small datasets, use k-fold cross-validation.
</details>
