````python
# ─────────────────────────────────────────────────────────────
# Setup & Imports
# ─────────────────────────────────────────────────────────────
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns

%matplotlib inline
np.random.seed(42)
sns.set_theme(style='whitegrid')

# Generate dataset — Indian city-wise sales data
n = 300
cities     = np.random.choice(['Mumbai', 'Bangalore', 'Delhi', 'Pune', 'Hyderabad'], n)
categories = np.random.choice(['Electronics', 'Clothing', 'Food', 'Books'], n)
sales      = np.random.randint(5000, 100000, n)
ratings    = np.random.uniform(2.0, 5.0, n).round(1)
month      = np.random.choice(range(1, 13), n)

sales_df = pd.DataFrame({
    'city': cities, 'category': categories,
    'sales': sales, 'rating': ratings, 'month': month
})

print('✅ Setup complete. Dataset shape:', sales_df.shape)
sales_df.head()
````

## 📖 Section A — Matplotlib Fundamentals

Every Matplotlib figure has two components:
- **Figure** — the outer container (`plt.figure(figsize=(w, h))`)
- **Axes** — the actual plot area (`ax = fig.add_subplot()` or `fig, ax = plt.subplots()`)

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y, color='steelblue', linewidth=2, label='Revenue')
ax.set_title('Monthly Revenue')
ax.set_xlabel('Month')
ax.set_ylabel('Revenue (₹)')
ax.legend()
plt.tight_layout()
plt.show()
```

> 💡 **Key Rule:** Always use `plt.tight_layout()` before `plt.show()` — it prevents labels from overlapping.

````python
# ─────────────────────────────────────────────────────────────
# Demo: Line Plot — Monthly sales trend
# ─────────────────────────────────────────────────────────────
monthly_avg = sales_df.groupby('month')['sales'].mean().reset_index()

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(monthly_avg['month'], monthly_avg['sales'],
        marker='o', color='steelblue', linewidth=2, markersize=6)
ax.fill_between(monthly_avg['month'], monthly_avg['sales'], alpha=0.15, color='steelblue')
ax.set_title('Average Monthly Sales Across Indian Cities', fontsize=14)
ax.set_xlabel('Month')
ax.set_ylabel('Average Sales (₹)')
ax.set_xticks(range(1, 13))
ax.set_xticklabels(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.tight_layout()
plt.show()
````

## 📖 Section B — Seaborn Statistical Plots

Seaborn is built on Matplotlib but designed for statistical data — it understands DataFrames natively.

```python
sns.boxplot(data=df, x='category', y='sales', palette='Set2')
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
sns.pairplot(df[['col1','col2','col3']], diag_kind='kde')
sns.histplot(df['sales'], bins=30, kde=True)
sns.scatterplot(data=df, x='rating', y='sales', hue='city')
```

> 💡 **Key Rule:** Seaborn is for **exploring distributions and relationships**. Use it for EDA. For production-quality custom charts, drop back to Matplotlib's low-level API.

````python
# ─────────────────────────────────────────────────────────────
# Demo: Seaborn — Boxplot and Correlation Heatmap
# ─────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Boxplot: sales distribution by category
sns.boxplot(data=sales_df, x='category', y='sales', palette='Set2', ax=axes[0])
axes[0].set_title('Sales Distribution by Product Category')
axes[0].set_ylabel('Sales (₹)')

# Heatmap: correlation between numeric columns
corr = sales_df[['sales','rating','month']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', ax=axes[1], center=0)
axes[1].set_title('Feature Correlation')

plt.tight_layout()
plt.show()
````

````python
# @title ✅ Solution — Exercise 4.3 (click to expand)

def visualise_dataset(df, target_col):
    stats = {}
    numeric_df   = df.select_dtypes(include='number')
    cat_cols     = df.select_dtypes(include='object').columns.tolist()
    numeric_features = [c for c in numeric_df.columns if c != target_col]

    n_plots = 1 + (1 if numeric_features else 0) + len(cat_cols)
    fig, axes = plt.subplots(1, n_plots, figsize=(6 * n_plots, 5))
    if n_plots == 1: axes = [axes]
    ax_idx = 0

    # 1. Target distribution
    sns.histplot(df[target_col], bins=30, kde=True, ax=axes[ax_idx], color='steelblue')
    axes[ax_idx].set_title(f'Distribution of {target_col}')
    stats['mean']     = df[target_col].mean()
    stats['std']      = df[target_col].std()
    stats['skewness'] = df[target_col].skew()
    ax_idx += 1

    # 2. Top correlations
    if numeric_features:
        corr = numeric_df.corr()[target_col].drop(target_col).abs().sort_values(ascending=False).head(5)
        axes[ax_idx].barh(corr.index, corr.values, color='darkorange')
        axes[ax_idx].set_title(f'Top Correlations with {target_col}')
        stats['top_correlations'] = corr.to_dict()
        ax_idx += 1

    # 3. Boxplots for categoricals
    for cat in cat_cols:
        if ax_idx < len(axes):
            sns.boxplot(data=df, x=cat, y=target_col, palette='Set2', ax=axes[ax_idx])
            axes[ax_idx].set_title(f'{target_col} by {cat}')
            axes[ax_idx].tick_params(axis='x', rotation=30)
            ax_idx += 1

    plt.suptitle(f'EDA Dashboard — Target: {target_col}', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
    return stats

stats = visualise_dataset(sales_df, 'sales')
print('Stats:', {k: round(v, 2) if isinstance(v, float) else v for k, v in stats.items() if k != 'top_correlations'})
assert 'mean' in stats
print('✅ Solution verified!')
````

## 🎤 Interview Q&A

<details>
<summary><strong>Q1: When should you use a bar chart vs histogram?</strong></summary>

**Answer:** Bar charts are for **categorical data** — discrete groups where each bar represents a category (e.g., sales by city). Histograms are for **continuous numeric data** — they show the frequency distribution of a variable by binning it into intervals (e.g., salary distribution). Key rule: categorical data → bar chart; distribution of one numeric variable → histogram.
</details>

<details>
<summary><strong>Q2: What does a boxplot show?</strong></summary>

**Answer:** A boxplot shows: median (middle line), Q1–Q3 range (box), 1.5×IQR fence (whiskers), and outliers (dots beyond the whiskers). The box spans the interquartile range (IQR = Q3−Q1). A box that is shifted high = high median. A long upper whisker = right-skewed data. Dots = outliers. Boxplots are excellent for comparing distributions across groups.
</details>

<details>
<summary><strong>Q3: What is the difference between Figure and Axes in Matplotlib?</strong></summary>

**Answer:** A Figure is the entire canvas — the window containing everything. An Axes is a single plot panel inside the Figure. One Figure can contain multiple Axes (subplots). Use `fig, ax = plt.subplots()` for one plot, `fig, axes = plt.subplots(2, 3)` for a 2×3 grid. Always call methods on `ax` (e.g., `ax.set_title()`), not on `plt`, when working with multiple subplots — otherwise you'll accidentally modify the wrong panel.
</details>

<details>
<summary><strong>Q4: When would you use a heatmap?</strong></summary>

**Answer:** Heatmaps are best for visualising a **matrix of values** — most commonly a correlation matrix, confusion matrix, or pivot table. They use colour intensity to encode value magnitude, making it easy to spot patterns (e.g., which features are strongly correlated). Always include `annot=True` so viewers can read exact values, and set `center=0` for correlation matrices so white = 0.
</details>
